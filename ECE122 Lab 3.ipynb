```python
# Q6 [2pts]: Let us consider a string that contains a binary number, complete the program using a for loop
#            to convert this binary number to a decimal number. As reminder of how this conversion operates,
#            the above figure is an example on how to obtain the number 37

bit=input("Enter a binary number: ")

bit_list=[i for i in bit]
reversed_bits=[]
num=int()
for j in bit_list:
  reversed_bits=[j]+reversed_bits
reversed_bits=[int(p) for p in reversed_bits] 
for k in range(len(reversed_bits)):
  if reversed_bits[k]==int(1):
    num=num+(2**k) 
  else:
    num=num+0  

# to complete


print("number is ", num)
```

```python
#Q7 [1pt]: This is the Triangular Number Sequence: 1,3,6,10,15,21,28,36,45,...
# which can be obtained using the math Formula: T(n)=T(n-1)+n and T(0)=0
# if you apply this formula, you will get T(1)=1, T(2)=3, T(3)=6, T(4)=10,...
# Using a while loop, write all the triangular numbers with T(n) smaller than 145
# You can use either a traditional while loop or a while True loop
# The output should look like (output is incomplete here)
""" 
T(1)=1
T(2)=3
T(3)=6
T(4)=10
T(5)=15
T(6)=21
T(7)=28
T(8)=36
T(9)=45
...
"""
n=1
x=1
while True:
  if x==1 and n==1:
    print("T(%s)=%s"%(n,x))
  x=x+n
  x=x+1
  n=n+1
  if x<145:
    print("T(%s)=%s"%(n,x))
  else:
    break  
# to complete
```
